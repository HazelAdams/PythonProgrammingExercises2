# Constant Menu

```python
#constants
VALIDCHOICES = "ABCE"

def choiceA():
    print("you choose option 1")

def choiceB():
    print("you choose option 2")
    
def choiceC():
    print("you choose option 3")
    
def choiceE():
    print("you have chosen to exit... goodbye!")
    
def menu():
    print("------MENU------")
    print("A) choice 1")
    print("B) choice 2")
    print("C) choice 3")
    print("E) exit")
    choice = input("please enter your choice ").upper()
    return choice

#main program starts here ---------

#show the menu
userChoice =menu()
if userChoice not in VALIDCHOICES:
    print("that was not a valid menu choice ")
elif userChoice == "A":
    choiceA()
elif userChoice == "B":
    choiceB()
elif userChoice == "C":
    choiceC()
elif userChoice == "E":
    choiceE()
    
#assume the users choice is NOT valid
valid = False
#keep asking until they make a valid choice
while not valid:
#show the menu
    userChoice =menu()
    if userChoice not in VALIDCHOICES:
        print("that was not a valid menu choice, try again")
    else:
        valid = True #so the loop doesnt repeat
        #handle the valid choice
        if userChoice == "A":
            choiceA()
        elif userChoice == "B":
            choiceB()
        elif userChoice == "C":
            choiceC()
        elif userChoice == "E":
            choiceE()

#assume the users choice is NOT valid
valid = False
#assume the user wants to choose a menu option
exit = False
#keep asking until they make a valid choice
while not valid or not exit:
#show the menu
    userChoice =menu()
    if userChoice not in VALIDCHOICES:
        print("that was not a valid menu choice, try again")
    else:
        valid = True #so the loop doesmt repeat
        #handle the valid choice
        if userChoice == "A":
            choiceA()
        elif userChoice == "B":
            choiceB()
        elif userChoice == "C":
            choiceC()
        elif userChoice == "E":
            choiceE()
            exit = True #so the loop doesnt repeat
    
```
